#**3.Under-Sampling**

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

# Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

# **4.Splitting the data into Features & Targets**

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

# **5.Split the data into Training data & Testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# 6.Model Training
 6.1 Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
model.fit(X_train, Y_train)
```

## 💡 What's happening here?

### ✅ 1. `from sklearn.preprocessing import StandardScaler`

This imports `StandardScaler`, a tool that standardizes your dataset's features (i.e., input values) by removing the mean and scaling them to unit variance:

image.png

This is important because many machine learning models (like logistic regression) converge faster and perform better when input features are on a similar scale.

---

### ✅ 2. `from sklearn.linear_model import LogisticRegression`

This imports the **Logistic Regression** model from scikit-learn.

---

### ✅ 3. `from sklearn.pipeline import make_pipeline`

`make_pipeline` allows you to chain multiple steps together—here, it's scaling the data and then passing it to logistic regression.

---

### ✅ 4. `model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))`

This creates a pipeline object that:

- First applies `StandardScaler` to `X_train`.
- Then fits a `LogisticRegression` model using the scaled data.
- `max_iter=1000` gives the model more time to converge than the default (100).

---

### ✅ 5. `model.fit(X_train, Y_train)`

This trains the pipeline:

- It scales `X_train`.
- Then it fits the logistic regression model on the scaled data.

All of this happens in **one call**, making your code cleaner and less error-prone.

---

## 🤖 Why use a pipeline?

- **Cleaner code**: No need to separately transform `X_train` and `X_test`.
- **Less risk of mistakes**: You won’t accidentally forget to scale `X_test` later.
- **Easier model deployment**: Everything (scaling + model) is packaged together.


#Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
